```python
# import seaborn library 
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset 
data = pd.read_csv('/microDec/data/final_results/final_resutls_91apps_bert_gpt.csv') 
#data = pd.read_csv('/microDec/data/final_results/final_resutls_91apps_2baselines.csv') 

  
# view the dataset 
print(data.head(5))
```

```python
plt.rcParams['font.family'] = 'serif'

plt.rcParams['font.size'] = 14  
plt.rcParams['font.weight'] = 'bold' 
x_values = list(range(20, 631, 75)) + [float('inf')]  # classes > 705

# Labels for x-axis
x_labels = [f"{x_values[i]}-{x_values[i+1]-1}" for i in range(len(x_values)-2)] + ["620+"]

# Categorize data into bins
class_counts = pd.cut(data['classes'], bins=x_values, right=False, labels=x_labels).value_counts(sort=False)

# Plot using plt.bar()
plt.figure(figsize=(10, 6))

# Bar plot with adjusted bar width
plt.bar(class_counts.index, class_counts.values, color='#1b9e77')  

# 
plt.xlabel('Number of Classes (Ranges)',fontsize=14)
plt.ylabel('Number of Apps',fontsize=14)



# Show the plot
plt.xticks(rotation=45, ha="right",fontsize=14)  
plt.tight_layout()
plt.savefig('/Users/ahmedalsayed/downloads/filename.png', dpi=1200)
plt.show()
```

```python
df = pd.DataFrame(data)

df = pd.melt(df, id_vars=['name'], value_vars=['b_ccoh', 'b_ccop', 'o_ccoh', 'o_ccop'],
                    var_name='Metric_Type', value_name='Value')

#new column to separateBaseline and Our
df['Group'] = df['Metric_Type'].apply(lambda x: 'Baseline' if 'b_' in x else 'Our')

# Add a new column for simplified metric naming
df['Metric'] = df['Metric_Type'].apply(lambda x: x[2:])


plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', hue='Metric', data=df)

plt.title('Comparison of ccoh and ccop for Baseline and Our')
plt.ylabel('')
plt.xlabel('')
plt.legend(title='')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'App Name': ['PetClinic', 'FXML-POS', 'JForum'],
    'Size': [23, 85, 232],
    'Topic Modeling': [186.47, 293.51, 242.56],
    'MAGNET': [81.51, 144.87, 2983.89],
    'BERT': [35.40, 68.96, 253.40],
    'GPT': [101.29, 258.55, 769.56]
}

df = pd.DataFrame(data)

df_sorted = df.sort_values('Size')
plt.rcParams['font.family'] = 'serif' 
plt.rcParams['font.size'] = 14  
plt.rcParams['font.weight'] = 'bold'

plt.figure(figsize=(10, 6))

methods = ['Topic Modeling', 'MAGNET', 'BERT', 'GPT']

for method in methods:
    plt.plot(df_sorted['Size'], df_sorted[method], marker='o', label=method)

plt.xlabel('No. of Classes',fontsize=14)
plt.ylabel('Time (seconds)',fontsize=14)
plt.title('')
plt.legend()
plt.grid(True)
plt.savefig('/microDec/time.png', dpi=1200)
plt.show()
```
